```python
default_conf_dict = {'je_cache_oblivious': True, 'je_metadata_thp': 'disabled', 'je_trust_madvise': False, 'je_retain': True, 'je_dss': 'secondary', 'je_narenas': 160, 'je_oversize_threshold': 8388608, 'je_background_thread': False, 'je_max_background_threads': 40, 'je_dirty_decay_ms': 10000, 'je_muzzy_decay_ms': 10000, 'je_lg_extent_max_active_fit': 6, 'je_tcache': True, 'je_lg_tcache_max': 15, 'je_thp': 'default', 'je_percpu_arena': 'default'}
default_conf_dict = dict(sorted(default_conf_dict.items()))
default_conf_dict.update({'rss': 1719.5})
default_conf_df = pd.DataFrame([default_conf_dict])
default_conf_df
```

```python
# Get max and min RSS rows
max_row = df.loc[[df['rss'].idxmax()]]  # Keep as DataFrame
min_row = df.loc[[df['rss'].idxmin()]]
default_row = default_conf_df.iloc[0]

# Concatenate and transpose to create vertical comparison
comparison_df = pd.concat([min_row, default_conf_df]).T
comparison_df.columns = ['Min RSS', 'Default RSS']  # Rename columns
```

```python
comparison_df
```

```python
plt.figure(figsize=(10, 8))
plt.plot(df.index, df['rss'], marker='o')
plt.ylim(bottom=0)

min_rss = df['rss'].min()
vertical_space = abs(default_rss - min_rss)
higher_line = max(default_rss, min_rss)
lower_line = min(default_rss, min_rss)

offset = (plt.ylim()[1] - plt.ylim()[0]) * 0.02  # 2% of y-axis range

if vertical_space < offset * 2:
    plt.axhline(y=default_rss, color='orange', linestyle='--')
    plt.text(plt.xlim()[1], higher_line + offset, f'{"Default" if higher_line == default_rss else "Min"} RSS:\n{higher_line:.5f}', 
             verticalalignment='bottom', 
             horizontalalignment='left')
    
    plt.axhline(y=min_rss, color='green', linestyle='--')
    plt.text(plt.xlim()[1], lower_line - offset, f'{"Default" if lower_line == default_rss else "Min"} RSS:\n{lower_line:.5f}', 
             verticalalignment='top', 
             horizontalalignment='left')

# Customize the plot
plt.title('RSS vs Iteration #')
plt.xlabel('Iteration #')
plt.ylabel('RSS')
plt.grid(True)

# Display the plot
#plt.tight_layout()
plt.show()
```

```python
percent_diff = 100 * (default_rss - df['rss'].min()) / default_rss
percent_diff
```

```python
score_cols = df.columns[-1:].copy()

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols = [e for e in numeric_cols if e not in score_cols]
categorical_cols = df.select_dtypes(include=['bool', 'object']).columns.tolist()

# Plotting the graphs
score_col = 'rss'
for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    plt.scatter(df[col], df[score_col], marker='o')
    plt.title(f'Plot {col}')
    plt.xlabel(col)
    plt.ylabel(score_col)
    plt.grid(True)
    plt.show()

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    # df.boxplot(column=score_col, by=col, grid=False, patch_artist=False, showfliers=False)
    # plt.scatter(df[col], df[score_col], marker='o')
    sns.boxplot(x=col, y=score_col, data=df, color='lightgray', boxprops=dict(facecolor='none', edgecolor='black', linewidth=1), fliersize=0)
    sns.scatterplot(x=col, y=score_col, data=df, color='blue', alpha=0.6)
    plt.title(f'Plot {col}')
    plt.suptitle('')
    plt.xlabel(col)
    plt.ylabel(score_col)
    plt.grid(True)
    plt.show()
```
